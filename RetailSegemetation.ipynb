```python
# FIX: make df an independent copy
df = df.copy()

# 1. Missing indicators (so model knows value was missing)
df['education_missing'] = df['education_level'].isnull().astype(int)
df['income_missing'] = df['annual_income'].isnull().astype(int)
df['teen_missing'] = df['num_teenagers'].isnull().astype(int)

# 2. Impute categorical column
df['education_level'] = df['education_level'].fillna("Unknown")

# 3. Impute numeric columns with median
df['annual_income'] = df['annual_income'].fillna(df['annual_income'].median())
df['num_teenagers'] = df['num_teenagers'].fillna(df['num_teenagers'].median())

df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

df = df.copy()

le = LabelEncoder()

for col in df.select_dtypes(include=['object']).columns: #object -> string || categorical data
    df[col + "_encoded"] = le.fit_transform(df[col])

print(df.head(10))
```

```python
df = df.drop(columns=["marital_status","education_level"])
df.tail(20)
```

```python
from sklearn.feature_selection import VarianceThreshold
import pandas as pd

# Make a copy to keep things clean
df_fs = df.copy()

# ----- 1. Remove low-variance features -----
var_selector = VarianceThreshold(threshold=0.01)  # remove features with <1% variance
var_selector.fit(df_fs)

low_var_mask = var_selector.get_support()

df_lowvar = df_fs.loc[:, low_var_mask]

print("Removed features (low variance):")
print(df_fs.columns[~low_var_mask].tolist())
print("\nRemaining features:", df_lowvar.shape[1])
```

```python

import numpy as np

# ----- 2. Remove highly correlated features -----
corr_matrix = df_lowvar.corr().abs()

# Upper triangle of the correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Features to drop
to_drop = [column for column in upper.columns if any(upper[column] > 0.85)]

print("Highly correlated features removed:", to_drop)

df_corr = df_lowvar.drop(columns=to_drop)
```

```python
from sklearn.cluster import KMeans

# Use the FINAL cleaned features to form clusters
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(df_corr)

# Add clusters to df_corr
df_corr["Cluster"] = clusters
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

# Target
y = df_corr["Cluster"]

# Features (drop target!)
X = df_corr.drop(columns=["Cluster"])

# Select best k features
k = 3
selector = SelectKBest(score_func=f_classif, k=k)
X_new = selector.fit_transform(X, y)

selected_mask = selector.get_support()
selected_features = X.columns[selected_mask]

print("Selected Features (ANOVA F-test):")
print(selected_features)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Extract features
X = df_corr[["annual_income", "spend_wine", "spend_meat"]]
labels = df_corr["Cluster"]

# Create a color map with strong contrast
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']  # blue, orange, green, red

fig = plt.figure(figsize=(14, 10))
ax = fig.add_subplot(111, projection="3d")

# Plot each cluster separately for clarity
for cluster_id in sorted(df_corr["Cluster"].unique()):
    subset = df_corr[df_corr["Cluster"] == cluster_id]
    ax.scatter(
        subset["annual_income"],
        subset["spend_wine"],
        subset["spend_meat"],
        s=60,
        alpha=0.85,
        color=colors[cluster_id % len(colors)],
        label=f"Cluster {cluster_id}",
        edgecolor="k",
        linewidth=0.4
    )

# Styling
ax.set_xlabel("Annual Income", fontsize=12)
ax.set_ylabel("Spend on Wine", fontsize=12)
ax.set_zlabel("Spend on Meat", fontsize=12)
ax.set_title("Enhanced 3D Visualization of Clusters", fontsize=15)

# Rotate for better angle
ax.view_init(elev=25, azim=135)

plt.legend()
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Use only the selected features
X_selected = df_corr[["annual_income", "spend_wine", "spend_meat"]]

# Clusters
labels = df_corr["Cluster"]

score = silhouette_score(X_selected, labels)
print("Silhouette Score:", score)
```

```python
# =============================
#  SVM ONE-VS-ALL CLASSIFIER
# =============================

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Attach KMeans cluster labels (NOT PCA)
df_corr["Cluster"]

# Target
y = df_corr["Cluster"]

# Features (drop target)
X = df_corr.drop(columns=["Cluster"])

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# SVM OVA
svm_ova = OneVsRestClassifier(
    SVC(kernel="rbf", C=1.0, gamma="scale")
)

# Train
svm_ova.fit(X_train, y_train)

# Predict
y_pred = svm_ova.predict(X_test)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("SVM Confusion Matrix")
plt.show()
```
