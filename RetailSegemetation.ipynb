```python
# FIX: make df an independent copy
df = df.copy()

# 1. Missing indicators (so model knows value was missing)
df['education_missing'] = df['education_level'].isnull().astype(int)
df['income_missing'] = df['annual_income'].isnull().astype(int)
df['teen_missing'] = df['num_teenagers'].isnull().astype(int)

# 2. Impute categorical column
df['education_level'] = df['education_level'].fillna("Unknown")

# 3. Impute numeric columns with median
df['annual_income'] = df['annual_income'].fillna(df['annual_income'].median())
df['num_teenagers'] = df['num_teenagers'].fillna(df['num_teenagers'].median())

df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

df = df.copy()

le = LabelEncoder()

for col in df.select_dtypes(include=['object']).columns: #object -> string || categorical data
    df[col + "_encoded"] = le.fit_transform(df[col])

print(df.head(10))
```

```python
df = df.drop(columns=["marital_status","education_level"])
df.tail(20)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca.fit(df)
PCA_ds = pd.DataFrame(pca.transform(df), columns=(["col1","col2", "col3"]))
PCA_ds.describe()
```

```python

import matplotlib.pyplot as plt
x =PCA_ds["col1"]
y =PCA_ds["col2"]
z =PCA_ds["col3"]
#To plot
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(projection="3d")
ax.scatter(x,y,z)
ax.set_title("A 3D Projection Of Data In The Reduced Dimension")
plt.show()
```

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
print('Elbow Method to determine the number of clusters to be formed:')
Elbow_M = KElbowVisualizer(KMeans(), k=10)
Elbow_M.fit(PCA_ds)
Elbow_M.show()
```

```python

kmeans = KMeans(n_clusters=4, random_state=0)  # choose 4 clusters
PCA_ds['Cluster'] = kmeans.fit_predict(PCA_ds[["col1","col2","col3"]])
```

```python
from matplotlib import pyplot as plt
from matplotlib import colors

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(1,1,1, projection='3d')

# Make a colormap with 4 distinct colors
cmap = colors.ListedColormap(["#682F2F", "#008000", "#0000FF", "#B9C0C9"])

# x, y, z = your PCA components
ax.scatter(x, y, z, s=40, c=PCA_ds["Cluster"], cmap=cmap)

ax.set_xlabel("PCA 1")
ax.set_ylabel("PCA 2")
ax.set_zlabel("PCA 3")
ax.set_title("3D Scatter Plot of Clusters")

plt.show()
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(PCA_ds[["col1","col2","col3"]], PCA_ds['Cluster'])
print("Silhouette Score:", score)
# Closer to 1 → well-separated clusters, around 0 → overlapping, negative → bad assignment.
```
